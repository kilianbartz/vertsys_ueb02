```python
import matplotlib.pyplot as plt

x = [100, 500, 1000, 2000, 2500, 3000]
data = [14056, 366835, 1498634, 6034477, 9389993, 13436610]

messages_per_node = [a/b for a, b in zip(data, x)]

# Create the main figure and axis
fig, ax1 = plt.subplots()

# Plot data on the primary y-axis
color1 = 'tab:blue'
ax1.set_xlabel('Zahl der Aktoren')
ax1.set_ylabel('Ausgetauschte Nachrichten bis zur Terminierung', color=color1)
ax1.plot(x, data, 'o-', color=color1, label='Nachrichten')
ax1.tick_params(axis='y', labelcolor=color1)

# Create a secondary y-axis
ax2 = ax1.twinx()

# Plot messages per node on the secondary y-axis
color2 = 'tab:orange'
ax2.set_ylabel('Nachrichten pro Aktor', color=color2)
ax2.plot(x, messages_per_node, 'o-', color=color2, label='Nachrichten pro Aktor')
ax2.tick_params(axis='y', labelcolor=color2)

# Add a legend
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

plt.title('Ausgetauschte Nachrichten und Nachrichten pro Aktor')
plt.tight_layout()
plt.show()
```
